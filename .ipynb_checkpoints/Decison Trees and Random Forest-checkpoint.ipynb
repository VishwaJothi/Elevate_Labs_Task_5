```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,auc
%matplotlib inline
```

```python
df=pd.read_csv("heart.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isna().sum()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
x=df.drop("target",axis=1)
y=df["target"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

# Decision Trees

```python
dt=DecisionTreeClassifier(random_state=42)
dt.fit(x_train_scaled,y_train)
y_pred_dt=dt.predict(x_test_scaled)
print("Decison Tree Accuracy:",accuracy_score(y_test,y_pred_dt))
print(classification_report(y_test,y_pred_dt))
```

```python
plt.figure(figsize=(20,10))
plot_tree(dt,feature_names=x.columns,class_names=['No Disease','Disease'],filled=True,
          fontsize=8,max_depth=3)
plt.show()
```

```python
param_grid={'max_depth':list(range(1,21))}
grid_dt=GridSearchCV(DecisionTreeClassifier(random_state=42),param_grid,cv=5,scoring='accuracy')
grid_dt.fit(x_train_scaled,y_train)
print("Best max_depth:",grid_dt.best_params_)
print("Best CV score:",grid_dt.best_score_)
```

```python
best_dt=grid_dt.best_estimator_
y_pred_best_dt=best_dt.predict(x_test_scaled)
print("Tuned Decision Tree Accuracy:",accuracy_score(y_test,y_pred_best_dt))
print(confusion_matrix(y_test,y_pred_best_dt))
print(classification_report(y_test,y_pred_best_dt))
```

# Random Forest

```python
rf=RandomForestClassifier(n_estimators=200,random_state=42)
rf.fit(x_train_scaled,y_train)
y_pred_rf=rf.predict(x_test_scaled)
print("Random Forest Accuracy:",accuracy_score(y_test,y_pred_rf))
print(classification_report(y_test,y_pred_rf))
```

```python
importances=pd.Series(rf.feature_importances_,index=x.columns).sort_values(ascending=False)
importances.head(15)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=importances.values,y=importances.index)
plt.title("Features Imporatnces(Random Forest)")
plt.show()
```

# ROC Curve

```python
# Decision Tree
y_score_dt = best_dt.predict_proba(x_test_scaled)[:, 1]
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_score_dt)

# Random Forest
y_score_rf = rf.predict_proba(x_test_scaled)[:, 1]
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_score_rf)

plt.figure(figsize=(8,6))
plt.plot(fpr_dt, tpr_dt, label=f'Decision Tree (AUC = {auc(fpr_dt, tpr_dt):.3f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc(fpr_rf, tpr_rf):.3f})')

plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import joblib
joblib.dump(best_dt,"best_decision_tree.pkl")
joblib.dump(rf,"random_forest_model.pkl")
print("Model Saved Successfully!")
```
